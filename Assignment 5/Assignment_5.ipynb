```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.corpus import wordnet as wn
```

```python
word = "happy"

synonyms = set()

for syn in wn.synsets(word):
    for lemma in syn.lemmas():
        synonyms.add(lemma.name())

print("Synonyms:", synonyms)
```

```python
antonyms = set()

for syn in wn.synsets(word):
    for lemma in syn.lemmas():
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())

print("Antonyms:", antonyms)
```

```python
hypernyms = set()

for syn in wn.synsets(word):
    for hyper in syn.hypernyms():
        hypernyms.add(hyper.name())

print("Hypernyms:", hypernyms)
```

```python
def semantic_relations(word):

    synonyms = set()
    antonyms = set()
    hypernyms = set()

    for syn in wn.synsets(word):

        # Synonyms
        for lemma in syn.lemmas():
            synonyms.add(lemma.name())

            # Antonyms
            if lemma.antonyms():
                antonyms.add(lemma.antonyms()[0].name())

        # Hypernyms
        for hyper in syn.hypernyms():
            hypernyms.add(hyper.name())

    return synonyms, antonyms, hypernyms
```

```python
word = "good"

syn, ant, hyp = semantic_relations(word)

print("Word:", word)
print("Synonyms:", syn)
print("Antonyms:", ant)
print("Hypernyms:", hyp)
```

```python
text = "The happy dog is very good and friendly"

words = text.lower().split()

for w in words:
    syn, ant, hyp = semantic_relations(w)

    print("\nWord:", w)
    print("Synonyms:", list(syn)[:5])
    print("Antonyms:", list(ant)[:5])
    print("Hypernyms:", list(hyp)[:5])
```
